# Tf-idf example

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
texts = [
    "good movie", "not a good movie", "did not like", 
    "i like it", "good one"
]
# using default tokenizer in TfidfVectorizer
tfidf = TfidfVectorizer(min_df=2, max_df=0.5, ngram_range=(1, 2))
features = tfidf.fit_transform(texts)
pd.DataFrame(
    features.todense(),
    #get_feature_names() is deprecated; see source:
    #https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html
    columns=tfidf.get_feature_names_out()
)
```

document frequncies:

df(“good”) = 3 (docs 1, 2, 5)

df(“movie”) = 2 (docs 1, 2)

1-grams present in “good movie” are good and movie, each with TF = 1

ln(5/3) + ln(5/2)

```python
# modifcations
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
texts = [
    "good breakfast", "not a good breakfast", "did not like breakfast", 
    "i liked it", "good pancakes"
]
# using default tokenizer in TfidfVectorizer
tfidf = TfidfVectorizer(
    ngram_range=(1,1),
    min_df=1,         
    max_df=1.0,      
    use_idf=True,
    smooth_idf=False, 
    norm=None,         
    sublinear_tf=False
)
features = tfidf.fit_transform(texts)
pd.DataFrame(
    features.todense(),
    #get_feature_names() is deprecated; see source:
    #https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html
    columns=tfidf.get_feature_names_out()
)
```
